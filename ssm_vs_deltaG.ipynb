```python
import nupack as nu
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from seqwalk import design
import pylcs
import pandas as pd
import iqplot
import bokeh

bokeh.io.output_notebook()
c = nu.Model(material="dna", celsius=22, sodium=0.3, magnesium=0.00)
```

```python
def randseq(length):
    return "".join(np.random.choice(["A", "C", "T", "G"], length))
#     return "".join(np.random.choice(["A", "C", "T"], length))

def rc(seq):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    return "".join(complement.get(base, base) for base in reversed(seq))
```

```python
%%time

np.random.seed(314)

ssms = [] # longest common subsequence
dG = [] # delta G

# simulate 50,000 pairs of 25nt sequences
for i in range(5*10**4):
    s1 = randseq(25)
    s2 = randseq(25)
    ssms.append(pylcs.lcs_string_length(s1, s2))
    dG.append(nu.pfunc([s1, rc(s2)], c)[1]) # delta G
```

```python
d = {"s" : ssms, "delta G" : dG}
df = pd.DataFrame(d)
df
```

```python
p = iqplot.ecdf(data=df, q="delta G", cats="s", 
                order=list(range(2,11)), x_range=(-20, -8))
bokeh.io.show(p, )
```
